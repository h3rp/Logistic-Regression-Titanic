```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder() #encode string to label or categorial numeric
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df.head()
```

```python
y = df['Survived']
x = df.drop(columns='Survived')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #biar data lebih seragam 
x_scaled = scaler.fit_transform(x)
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, random_state=42, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty='elasticnet', max_iter=10000, solver='saga', l1_ratio=0.5) #dengan hyperparameter
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```
